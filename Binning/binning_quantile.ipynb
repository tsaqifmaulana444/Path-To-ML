### Binning dengan quantile
quantile = membagi data berdasarkan persentilnya

```python
import pandas as pd
import numpy as np
from sklearn.pipeline import Pipeline
```

```python
df = pd.read_csv('datasetgaji.csv')
df.head(6)
```

```python
df['gaji_binned_q'] = pd.qcut(df['Gaji Orang Tua'], 2)
df
```

```python
kategori = ['Kecil', 'Besar']
df['gaji_binned_q'] = pd.qcut(df['Gaji Orang Tua'], q=2, labels=kategori)
df
```
