Ekstrasi fitur tidak menghapus atribut yang kurang berpengaruh terhadap performa dataset, namun fitur ini menyatukan (meleburkan) semua fitur menjadi suatu kesatuan yang sama

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
%matplotlib inline

from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
df = pd.DataFrame(cancer["data"],columns=cancer["feature_names"])
df.head()
```

### PCA Visulalization (Principal Component Analyst)

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)

# data sebelum di scale
scaled_data.shape
```

```python
# data setelah di scale
x_pca.shape
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer["target"],cmap="plasma")
plt.xlabel('First principal component')
plt.ylabel('Second principal component')
```

### Interpreting The Component

```python
pca.components_
```

```python
df_comp = pd.DataFrame(pca.components_, columns=cancer['feature_names'])
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df_comp, cmap='plasma')
```
