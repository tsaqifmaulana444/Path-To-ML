# Supervised Learning
- Supervised learning adalah salah satu jenis pendekatan dalam pembelajaran mesin (machine learning) di mana model diajari untuk mengenali pola dalam data berdasarkan contoh-contoh yang telah diberikan sebelumnya, yang disebut sebagai "data latih" (training data). Dalam supervised learning, data latih yang digunakan memiliki label atau anotasi, yang berarti setiap contoh data latih memiliki informasi tentang hasil yang diinginkan atau kelas yang sesuai.

- Tujuan utama dari supervised learning adalah untuk membuat model yang dapat melakukan prediksi atau klasifikasi terhadap data baru yang belum dilihat sebelumnya. Model yang dihasilkan akan mempelajari pola-pola dalam data latih, dan kemudian menggunakannya untuk membuat prediksi atau klasifikasi terhadap data baru berdasarkan pola-pola yang telah dipelajari.

- Contoh umum dari aplikasi supervised learning meliputi pengenalan gambar (image recognition), klasifikasi spam email, prediksi harga rumah berdasarkan fitur-fitur tertentu, pengenalan suara (speech recognition), dan banyak lagi. 

# Unsupervised Learning
- Unsupervised learning adalah jenis pendekatan dalam pembelajaran mesin di mana model belajar untuk mengenali pola dalam data tanpa adanya label atau anotasi yang diberikan sebelumnya. Dalam unsupervised learning, model tidak diberikan informasi tentang hasil yang diinginkan atau kelas yang sesuai, melainkan dibiarkan untuk secara otomatis menemukan pola, struktur, atau representasi yang tersembunyi dalam data.

- Tujuan utama dari unsupervised learning adalah untuk menggali informasi yang berguna dari data yang tidak berlabel, seperti mengidentifikasi kelompok atau cluster data yang serupa berdasarkan kesamaan fitur, mengurangi dimensi data untuk memperoleh representasi yang lebih ringkas, atau menghasilkan data sintetis yang baru.

- Contoh umum dari aplikasi unsupervised learning meliputi klasifikasi dokumen, pengelompokan pelanggan berdasarkan perilaku pembelian, segmentasi pasar, deteksi anomali dalam data, penggalian pola asosiasi dalam data transaksi, atau pengurangan dimensi data dalam tugas visualisasi data.

# Reinforcement Learning
- Reinforcement learning adalah salah satu jenis pendekatan dalam pembelajaran mesin di mana agen (agent) belajar untuk mengambil tindakan (actions) dalam suatu lingkungan (environment) untuk mencapai tujuan tertentu. Dalam reinforcement learning, agen belajar melalui interaksi berulang dengan lingkungan, menerima umpan balik atau sinyal penguatan (rewards) berdasarkan tindakan yang diambil, dan memperbarui keputusannya untuk memaksimalkan penerimaan penguatan dalam jangka panjang, Agen juga dapat belajar dari kesalahan yang telah agen lakukan

- Tujuan utama dari reinforcement learning adalah untuk menghasilkan kebijakan (policy) yang optimal, yaitu keputusan tentang tindakan apa yang harus diambil dalam setiap situasi untuk mencapai tujuan yang diinginkan. Reinforcement learning cocok untuk masalah di mana tidak ada data latih yang diberikan sebelumnya, dan agen harus belajar melalui trial and error untuk mengoptimalkan tindakan yang diambil berdasarkan umpan balik yang diterima dari lingkungan.

- Contoh umum dari aplikasi reinforcement learning meliputi pengendalian robot, permainan komputer, pengaturan lalu lintas cerdas, sistem rekomendasi, manajemen sumber daya dalam jaringan komputer, dan banyak lagi.

# Data Diskrit
- Data diskrit adalah jenis data yang hanya bisa mengambil nilai-nilai tertentu atau terpisah dalam suatu rentang nilai. Contohnya adalah jumlah siswa dalam sebuah kelas yang hanya bisa berupa angka bulat, seperti 25 atau 30, dan tidak bisa mengambil nilai desimal di antara kedua nilai tersebut.

# Data Kontinu
- Data kontinu adalah jenis data yang dapat mengambil nilai di antara suatu rentang nilai dengan tak terbatas jumlahnya. Contohnya adalah tinggi badan seseorang yang bisa memiliki nilai seperti 160.5 cm atau suhu dalam suatu daerah yang bisa mencapai 25.75 derajat Celsius. Dalam data kontinu, ada tak terhingga jumlah nilai di antara dua nilai tertentu.

Dalam ringkasannya, data diskrit terdiri dari nilai-nilai terpisah dan hanya bisa mengambil nilai-nilai bulat, sedangkan data kontinu dapat mengambil nilai-nilai di antara rentang nilai dan bisa termasuk nilai desimal.
